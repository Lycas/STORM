```python
import graphlab as gl
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider, Button, RadioButtons
import os
import graphlab.aggregate as agg
import numpy as np
import seaborn as sns
```

```python
Raw_Data = gl.SFrame.read_csv('160519 -- V 001.txt', delimiter = '	',header = True)
Sauce = gl.SFrame({'X': Raw_Data['Xc']-min(Raw_Data['Xc']),'Y': Raw_Data['Yc']-min(Raw_Data['Yc']),'T':Raw_Data['Frame']})
```

```python
Sauce = Sauce.sort('T')
Sauce['Z']=0
Datas = Sauce.to_numpy()
Totals = gl.SArray(data = [],dtype = int)
Delta = 25

for i in range(0,len(Datas[:,0])-101):
    count = 0
    if Datas[i,3] == 0:
        for j in range(i+1,i+100):
            if Datas[j,1] < (Datas[i,1] + Delta) and Datas[j,1] > (Datas[i,1] - Delta) and Datas[j,2] < (Datas[i,2] +Delta) and Datas[j,2] > (Datas[i,2] - Delta):
                count = count + 1
                Datas[j,3] = Datas[j,0]-Datas[i,0]
        sfn = gl.SArray(data = [count],dtype = int)
        Totals = Totals.append(sfn)
Dist = gl.SFrame({'Blinks':Totals})
Dist = Dist.to_dataframe()
```

```python
sns.distplot(Dist.head(1000),color = 'red', label="Blinks", rug=True, kde_kws={"bw":1, "shade":True, "shade_lowest":False})
plt.show()
```

```python
New_Pts = gl.SFrame({'Frame':Datas[:,0],'X':Datas[:,1],'Y':Datas[:,2],'Count':Datas[:,3]})
```

```python
OG_Pts = New_Pts['Count'] == 0
```

```python


#Single locals
Single_Local = New_Pts[OG_Pts].to_dataframe()
g = sns.jointplot(x='X',y= 'Y', data=Single_Local, kind="kde",bw=90,cmap="Reds",shade_lowest=False,shade=True)
g.plot_joint(plt.scatter, c="b", s=5, linewidth=1, marker="+")
plt.figure(1)

#Raw Locals
All_Local = New_Pts.to_dataframe()
g = sns.jointplot(x='X',y= 'Y', data=All_Local, kind="kde",bw=90,cmap="Reds",shade_lowest=False,shade=True)
g.plot_joint(plt.scatter, c="b", s=5, linewidth=1, marker="+")
plt.figure(2)


plt.show()
```
